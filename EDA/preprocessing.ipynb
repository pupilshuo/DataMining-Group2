```python
import pandas as pd
import csv
import timeit
```

```python
data = pd.read_csv('order_products.csv')
data.head(10)
```

```python
data.shape
```

```python
df = data[['order_id','product_id','product_name']]
```

```python
df.head(10)
```

```python
#df = df.iloc[0:10,:]
```

```python
df.shape
```

```python
df['product_id'] = df['product_id'].astype('str')
```

```python
def concat_id(x):
    return pd.Series({
        'product_id':','.join(x['product_id'].unique())
    }
    )

def concat_name(x):
    return pd.Series({
        'product_name':','.join(x['product_name'].unique())
    }
    )

start = timeit.default_timer()
product_id = df.groupby(df['order_id']).apply(concat_id).reset_index()
stop = timeit.default_timer()
print('Time: ', stop - start)  
start = timeit.default_timer()
product_name = df.groupby(df['order_id']).apply(concat_name).reset_index()
stop = timeit.default_timer()
print('Time: ', stop - start)  
```

```python
product_id.head(20)
```

```python
product_name.head(20)
```

```python
product_id = product_id[['product_id']]
product_name = product_name[['product_name']]
```

```python
product_id.head(20)
```

```python
product_name.head(20)
```

```python
id_ls = product_id.values.tolist()
name_ls = product_name.values.tolist()
```

```python
all_id = []
all_name = []

for item in id_ls:
    all_id.append(item[0].split(','))
for item in name_ls:
    all_name.append(item[0].split(','))
```

```python
all_id
```

```python
arow = len(all_id)
acol = len(all_id[0])
print("Rows : " + str(arow))
```

```python
all_name
```

```python
arow = len(all_name)
acol = len(all_name[0])
print("Rows : " + str(arow))
```

```python
with open('product_id_set.csv', 'w', newline='') as f:
    write = csv.writer(f)
    write.writerows(all_id)
print('id written')
```

```python
with open('product_name_set.csv', 'w', newline='', encoding="utf-8") as f:
    write = csv.writer(f)
    write.writerows(all_name)
print('name written')
```
